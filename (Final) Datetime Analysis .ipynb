```python
import pandas as pd
import datetime
```

```python
df=pd.read_csv("crimechicago2.csv") #index_col='Date'
df.tail()
```

```python
df['Date'] = pd.to_datetime(df["Date"]) #df1 = df.iloc[: 500,] to load faster
```

# Date vs. Primary Type

```python
#Finding the correlation between the dates and the frequency of the crime type,Let's focus reading the dataset based on the dates vs. the primary as this will be the main focus of the project's analysis

df_crimetype = df[["Date","Primary Type"]]
```

```python
crimes_2013= df_crimetype[df_crimetype["Date"].isin(pd.date_range("2013-01-01", "2013-12-31"))]
#pd.DataFrame(pd.date_range(start='2013-01-01', end='2013-12-31', freq='D'), columns = ['Date'])
crimes_2013.tail()
```

```python
crimes_2013['weekday']=crimes_2013["Date"].dt.weekday
crimes_2013.head()
```

# convert date format to weekday

```python
crimecount_2013 = crimes_2013.groupby(["weekday","Primary Type"]).size()
crimecount_2013
```

```python
crimelist_2013 = crimes_2013[['Primary Type','weekday']].values.tolist()
crimelist_2013
```

# Applying Apriori algorithm on crime types for the year 2013

```python
from apyori import apriori

results = list(apriori(crimelist_2013))
for r in results:
    print('Days: ' + str(r.items))
    print('Support: ' + str(r.support))
    print('Confidence: ' + str(r.ordered_statistics[0].confidence) + '\n')
```

# Time vs. Crime Type

convert datetime to time and time categorized to day, noon,night, late night

```python
def time_of_day(timestamp):
    hr = timestamp.hour
    if hr>=6 and hr<12: return "daylight"
    elif hr>=12 and hr<18: return "afternoon"
    elif hr>=18 and hr<24: return "night"
    else: return "late night"
```

```python
df_crimetype['time_of_day'] = df_crimetype['Date'].apply(time_of_day)
```

```python
df_crimetype.head()
```

```python
df_crimetimetype = df_crimetype[['time_of_day','Primary Type']].values.tolist()#.astype(str)
df_crimetimetype[:5]
```

```python
from apyori import apriori

results = list(apriori(df_crimetimetype ,min_support=0.05))
for r in results:
    print('Days: ' + str(r.items))
    print('Support: ' + str(r.support))
    print('Confidence: ' + str(r.ordered_statistics[0].confidence) + '\n')
```
