```python
import pandas as pd
df=pd.read_csv("crimechicago2.csv") #index_col='Date'
df.tail(1)
```

# Ward vs. Primary Type

```python
df_wardtype = df[["Ward","Primary Type"]]
df_wardtype["Ward"]=df_wardtype["Ward"].astype(str)
df_wardtype.dtypes
```

```python
df_wardcount=df_wardtype.groupby(["Ward","Primary Type"]).size()
df_wardcount.head()
```

```python
df_wardtype = df_wardtype[["Ward","Primary Type"]].values.tolist()
df_wardtype
```

```python
from apyori import apriori

results = list(apriori(df_wardtype,min_support=0.01))
for r in results:
    print('Days: ' + str(r.items))
    print('Support: ' + str(r.support))
    print('Confidence: ' + str(r.ordered_statistics[0].confidence) + '\n')
```

# Can we find whether beat is more prone to crime on Monday?

```python
df_beatwardtyp= df[["Beat","Ward","Primary Type"]]
df_beatwardtyp.head()
```

```python
df_beatwardtypcount=df_beatwardtyp.groupby(["Beat","Ward","Primary Type"]).size()
df_beatwardtypcount
```

```python
df_beatwardtype = df[["Beat","Ward","Primary Type"]].values.tolist()
df_beatwardtype
```
